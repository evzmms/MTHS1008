# Exercises: Problem Sheet 1

1. Using Gaussian elimination, find the solutions to the following systems of equations

\begin{eqnarray*}
(\textrm{a})\qquad &  \phantom{-}x &+& 6y &+&  z &=&  \phantom{-}0 &\qquad\qquad& (\textrm{b}) \qquad & x &-&  y &+&  z &=&  \phantom{-}1 \\
             &  \phantom{-}x &-& 4y &+& 2z &=& -2 &&              & x &+&  y &&     &=&  \phantom{-}6 \\
             &            -x &-& 2y &+&  z &=&  \phantom{-}2 &&              & x &-& 5y &+& 3z &=& -9
\end{eqnarray*}

2. Using Gaussian elimination, find the solutions to the following systems of equations

\begin{eqnarray*}
(\textrm{a})\qquad &  2x &+& 3y &-&  z &=& 0 &\qquad\qquad& (\textrm{b}) \qquad & 4x &+&  y &+& 5z &=&  0 \\
             &  x &-& 2y &+& 5z &=& 0 &&              & \phantom{4}x &+&  2y &-& 4z  &=&  0 \\
             &  3x &-& y &-& 2z &=& 0 &&              & -x &-& y &+& \phantom{4}z &=& 0
\end{eqnarray*}

[**Hint**: clearly $x = y = z = 0$ is a solution to both systems of equations. Using Gaussian elimination you should be able to say whether either system of equations has any other solutions, besides $x = y = z = 0$, or whether this solution is unique.]

4. Given $\displaystyle {\bf{A}} = \left(\begin{array}{rr} 1 & 1 \\ 2 & -1 \\ 0 & 3\end{array}\right)$, $\displaystyle {\bf{B}} = \left(\begin{array}{rr} 0 & -3 \\ -1 & 2 \end{array}\right)$, and $\displaystyle {\bf{C}} = \left(\begin{array}{rr} 0 & -5 \\ 3 &  2 \end{array}\right)$:

* evaluate $\left({\bf{A}}{\bf{B}}\right){\bf{C}}$ and ${\bf{A}}\left({\bf{B}}{\bf{C}}\right)$;

* verify that ${\bf{A}}\left({\bf{B}} + {\bf{C}}\right) = {\bf{A}}{\bf{B}} + {\bf{A}}{\bf{C}}$;

* find the matrix ${\bf{M}}$ such that ${\bf{B}}^3 + 3{\bf{C}}^2 + 2{\bf{M}} = {\bf{0}}$.

8. Show that the determinant $\displaystyle \left|\begin{array}{rrr} 4 & -3 & 1 \\ 7 & 2 & -2 \\ 1 & -5 & 3 \end{array}\right|$ has the value 16 by:

* expanding along the third row;

* expanding along the second column.

9. Show that the determinant of $\displaystyle {\bf{A}} = \left(\begin{array}{ccc} t+3 & -1 & 1 \\ 5 & t-3 & 1 \\ 6 & -6 & t+4 \end{array}\right)$ is $\left(t + 2\right)\left(t - 2\right)\left(t + 4\right)$.

[**Hint**: notice that the answer is nicely factorized.  What method might you use to produce an answer in such a form?]

10. Write each of the following systems of equations as an augmented matrix, and solve each system by applying elementary row operations to the augmented matrix.

(a)

\begin{eqnarray*}
x\ &-& 2y &+& 2z &=& -3 \\
2x\ &+& \phantom{2}y &-& 3z &=& \phantom{-}8 \\
-x\ &+& 3y &+& 2z &=& -5
\end{eqnarray*}

(b)

\begin{eqnarray*}
x\ &-& 2y &+& 2z &=& -3 \\
2x\ &+& \phantom{2}y &-& 3z &=& \phantom{-}8 \\
x\ &+& 8y &-& 12z &=& \,25
\end{eqnarray*}

(c)

\begin{eqnarray*}
x\ &-& 2y &+& 2z &=& -3 \\
2x\ &+& \phantom{2}y &-& 3z &=& \phantom{-}8 \\
9x\ &-& 3y &-& 3z &=& \phantom{-}\alpha
\end{eqnarray*}

&emsp;&emsp;&nbsp;For what value of $\alpha$ is there a non-trivial solution to (c)?

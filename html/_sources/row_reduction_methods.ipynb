# Row Reduction Methods

* The technique introduced in Week 2 (Introduction to Linear Systems, Further Linear Systems: 3 x 3 Systems) for solving systems of equations "Gaussian Elimination" can be applied using matrices

* We note that the full equations need not be written out at each stage, only the numerical coefficients are required

* These coefficients can be written in matrix form

## The Augmented Matrix

Suppose we wish to solve the following system of equations for unknowns $x_1$, $x_2$, and $x_3$

\begin{eqnarray*}
3x_1 \,&-&\, \phantom{3}x_2 \,&+&\, 2x_3 \,&=&\, \phantom{-}3 \\
2x_1 \,&+&\, \phantom{3}x_2 \,&+&\, \phantom{2}x_3 \,&=&\, -2 \\
x_1 \,&-&\, 3x_2 && \,&=&\, \phantom{-}5
\end{eqnarray*}

In matrix form this is ${\bf{A}}\boldsymbol{x}=\boldsymbol{b}$, where

\begin{equation*}
{\bf{A}} = \left(\begin{array}{rrr} 3 & -1 & 2 \\ 2 & 1 & 1 \\ 1 & -3 & 0 \end{array}\right),
\quad
\boldsymbol{x} = \left(\begin{array}{l} x_1 \\ x_2 \\ x_3 \end{array}\right),
\quad
\boldsymbol{b} = \left(\begin{array}{r} 3 \\ -2 \\ 5 \end{array}\right).
\end{equation*}

:::{prf:definition} Augmented Matrix

We write down the **augmented matrix** $\left[\bf{A}|\boldsymbol{b}\right]$ of coefficients and right-hand sides

\begin{equation*}
\left[\begin{array}{rrr|r} 3 & -1 & 2 & 3 \\ 2 & 1 & 1 & -2 \\ 1 & -3 & 0 & 5 \end{array}\right]
\end{equation*}

:::

### Gaussian Elimination

* **Step 1**  
We write down the augmented matrix of coefficients and right-hand sides

\begin{equation*}
\left[\begin{array}{rrr|r} 3 & -1 & 2 & 3 \\ \color{red}2 & 1 & 1 & -2 \\ \color{red}1 & \color{red}-3 & 0 & 5 \end{array}\right]
\end{equation*}

&emsp;&emsp;&nbsp;We use **row operations** to carry out Gaussian elimination, making the terms below the leading diagonal 

&emsp;&emsp;&nbsp;(those in red) equal to zero.

&emsp;&emsp;&nbsp;We do not need to write down equations for $x_1$, $x_2$, and $x_3$ at each stage, we may simply deal with the 

&emsp;&emsp;&nbsp;numerical coefficients in the augmented matrix.

* **Step 2**  
Remove the terms below the leading diagonal in the first column

\begin{equation*}
\left[\begin{array}{rrr|r} 3 & -1 & 2 & 3 \\ \color{red}2 & 1 & 1 & -2 \\ \color{red}1 & -3 & 0 & 5 \end{array}\right]
\end{equation*}

\begin{equation*}
\left\{\begin{array}{l}
\color{gray}R_1\rightarrow R_1 \\
R_2 \rightarrow 3R_2 - 2R_1 \\
R_3 \rightarrow 3R_3 - R_1
\end{array}\right. \qquad\qquad
\left[\begin{array}{rrr|r} 3 & -1 & 2 & 3 \\ \color{red}0 & 5 & -1 & -12 \\ \color{red}0 & -8 & -2 & 12
\end{array}\right]
\end{equation*}

* **Step 3**  
Remove the terms below the leading diagonal in the first column

\begin{equation*}
\left[\begin{array}{rrr|r} 3 & -1 & 2 & 3 \\ 0 & 5 & -1 & -12 \\ 0 & \color{red}-8 & -2 & 12
\end{array}\right]
\end{equation*}

\begin{equation*}
\left\{\begin{array}{l}
\color{gray}R_1\rightarrow R_1 \\
\color{gray}R_2\rightarrow R_2 \\
R_3 \rightarrow 5R_3 + 8R_2
\end{array}\right. \qquad\qquad
\left[\begin{array}{rrr|r} 3 & -1 & 2 & 3 \\ 0 & 5 & -1 & -12 \\ 0 & \color{red}0 & -18 & -36
\end{array}\right]
\end{equation*}

* **Step 4**
Rewrite each row as an equation for $x_1$, $x_2$, and $x_3$.  
The last row gives us an equation for $x_3$:

\begin{equation*}
-18x_3 = -36.
\end{equation*}

&emsp;&emsp;&nbsp;Solving this equation gives $x_3 = 2$.

* **Step 5**
Substitute $x_3$ into the equation from row 2:

\begin{equation*}
5x_2 - x_3 = -12 \quad\Rightarrow\quad 5x_2 = -12 + 2 = -10 \quad\Rightarrow\quad x_2 = -2.
\end{equation*}

* **Step 6**
Substitute $x_2$ and $x_3$ into the equation from row 1:

\begin{equation*} 
3x_1 - x_2 + 2x_3 = 3 \quad\Rightarrow\quad 3x_1 = 3 - 2 - 4 = -3 \quad\Rightarrow\quad x_1 = -1.
\end{equation*}

The operations in Steps 2 and 3 are examples of **elementary row operations** (EROs).

:::{prf:definition} Elementary Row Operations

There are three types of elementary row operation

1. **interchange any two rows**;  
This is equivalent to swapping the order of any two equations.

2. **multiply any row by any nonzero constant**;  
This is equivalent to multiplying both sides of a given equation by the constant.

3. **add a multiple of one row to another**;  
This is equivalent to adding a multiple of one equation to another.

:::

:::{admonition} Warning

* It is important to distinguish EROs from the broader range of operations that may be applied to determinants

* In particular, **operations to columns are not allowed** when solving a system of equations by EROs.

:::

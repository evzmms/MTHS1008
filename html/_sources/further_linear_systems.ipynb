# Further Linear Systems: 3 x 3 Systems

## A 3 x 3 System

A similar procedure can be used to solve a system of three linear equations for three unknowns $x_1$, $x_2$, and $x_3$.

For example, suppose we wish to solve

\begin{eqnarray*}
2x_1 &+& 3x_2 &-&  x_3 &=& \phantom{-}5 \tag{1.8} \\
4x_1 &+& 4x_2 &−& 3x_3 &=& \phantom{-}3 \tag{1.9} \\
2x_1 &−& 3x_2 &+&  x_3 &=& −1 \tag{1.10}
\end{eqnarray*}

to obtain $x_1$, $x_2$, and $x_3$. There are more steps involved because there are more equations and more unknowns.

* **Step 1**  
Eliminate $x_1$ from equations (1.9) and (1.10) by subtracting multiples of (1.8).  
Replace  

\begin{align*}
(1.9) &\rightarrow (1.9)−2\times(1.8) \\
(1.10) &\rightarrow (1.10)−(1.8)
\end{align*}

&emsp;&emsp;&nbsp;to leave the system  
\begin{eqnarray*}
2x_1 &+& 3x_2 &-&  x_3 &=& \phantom{-}5 \tag{1.11} \\
     &−& 2x_2 &−&  x_3 &=& −7 \tag{1.12} \\
     &−& 6x_2 &+& 2x_3 &=& −6 \tag{1.13}
\end{eqnarray*}

* **Step 2**  
Eliminate $x_2$ from (1.13). To do this, we subtract an appropriate multiple of (1.12) from (1.13).  (If we subtract a multiple of (1.11) from (1.13) instead, then in the process of eliminating $x_2$ from (1.13) we reintroduce $x_1$!)

&emsp;&emsp;&nbsp;Replace  

$$(1.13) \rightarrow (1.13) −3\times(1.12)$$

&emsp;&emsp;&nbsp;to leave

\begin{eqnarray*}
2x_1 &+& 3x_2 &−& x_3 &=& \phantom{-}5 \tag{1.14} \\
     &−& 2x_2 &−& x_3 &=& −7 \tag{1.15} \\
     &&       && 5x_3 &=& \phantom{-}15 \tag{1.16}
\end{eqnarray*}

* **Step 3**  
Solve the "$1\times1$ system" (1.16) to find $x_3$.  In this case there is a unique solution to (1.16) 

$$x_3 = \frac{15}{5} = 3.$$

* **Step 4**  
Substitute this result back into (1.15) to give $x_2$ from 

$$−2x_2 − 3 = −7 \quad\Rightarrow\quad 2x_2 = 4 \quad\Rightarrow\quad x_2 = 2.$$

* **Step 5**  
Substitute the values we have found for $x_2$ and $x_3$ back into (1.14) to give

$$2x_1 + 6 − 3 = 5 \quad\Rightarrow\quad 2x_1 = 2 \quad\Rightarrow\quad x_1 = 1.$$

&emsp;&emsp;&nbsp;Finally, the solution to the system (1.8), (1.9), (1.10), is 

$$x_1 = 1, \quad x_2 = 2, \quad x_3 = 3.$$

&emsp;&emsp;&nbsp;This solution is **unique**.

:::{prf:definition} Gaussian Elimination
This method is known as **Gaussian Elimination** and can be applied to any system where the number of equations equals the number of unknowns.
:::

## Gaussian Elimination in 3 x 3 Systems

We apply Gaussian elimination to the equations

\begin{eqnarray*}
 x_1 &+&  x_2 &+& 2x_3 &=& 1 \tag{$R_1$} \\
2x_1 &−&  x_2 &−&  x_3 &=& 1 \tag{$R_2$} \\
 x_1 &+& 4x_2 &+& 7x_3 &=& 2 \tag{$R_3$}
\end{eqnarray*}
using more compact notation. 

* **Step 1**  
Eliminate $x_1$ from the last two equations

\begin{eqnarray*}
    &&&&                              & x_1 &+& ~x_2 &+& 2x_3 &=  1 \tag{$R_1'$}\\
R_2 &\rightarrow& R_2 &-& 2R_1 \qquad &     &−& 3x_2 &−& 5x_3 &= −1 \tag{$R_2'$}\\
R_3 &\rightarrow& R_3 &-& R_1  \qquad &     & & 3x_2 &+& 5x_3 &=   1 \tag{$R_3'$}
\end{eqnarray*}

* **Step 2**  
Eliminate $x_2$ from the last equation (use $R_2'$ so $x_1$  doesn't come back!)

\begin{eqnarray*}
    &&&&                              & x_1 &+& ~x_2 &+& 2x_3 &=  1 \tag{$R_1''$}\\
    &&&&                              &     &−& 3x_2 &−& 5x_3 &= −1 \tag{$R_2''$}\\
R_3' &\rightarrow& R_3' &+& R_2'   \qquad &&&&&          0x_3 &=  0 \tag{$R_3''$}
\end{eqnarray*}

* **Step 3**  
The last equation is a "$1\times1$ system" which is easily solved. In this case the equation

$$0 x_3=   0$$

&emsp;&emsp;&nbsp;is satisfied for any value of $x_3$ and there infinitely many solutions.

&emsp;&emsp;&nbsp;Let

$$x_3 = t,$$

&emsp;&emsp;&nbsp;where $t$ is an arbitrary number.

* **Step 4**  
Substitute $x_3$ back into the second equation $R_2''$ to get

$$−3x_2 − 5x_3 = −3x_2 − 5t = −1 \quad\Rightarrow\quad x_2 = \frac{1}{3} − \frac{5t}{3}.$$

* **Step 5**  
Substitute $x_2$ and $x_3$ back into the first equation $R_1''$ to get

$$x_1 + x_2 + 2x_3 = x_1 + \frac{1}{3} − \frac{5t}{3} + 2t = 1 \quad\Rightarrow\quad x_1 = \frac{2}{3} − \frac{t}{3}.$$

**Conclusion**  
For any given value of $t$

\begin{align*}
x_1 &= \frac{2}{3} - \frac{t}{3}, \\
x_2 &= \frac{1}{3} − \frac{5t}{3}, \\
x_3 &= t,
\end{align*}

is a solution of the original set of equations and there are an infinite number of such solutions.

:::{admonition} Exercise
Check this conclusion by direct substitution in the original equations R1, R2, and R3.
:::

## 3 x 3 Systems

* Equations of the form $a x_1 + b x_2 = u$ represent lines in two-dimensional space.
* Equations of the form $a x_1 + b x_2 + cx_3 = u$ represent planes in three-dimensional space.
* Systems of 3 equations in 3 unknowns represent 3 planes in three-dimensional space.  We still have the three possible outcomes of: unique solution, no solution, infinitely many solutions.

# Eigenvalues and Eigenvectors

Eigenvalue problems arise in many situations, for example

* calculating the natural frequencies of oscillation of a vibrating system

* finding principal axes of stress and strain

* calculating the oscillations of an electrical circuit

:::{prf:definition} Eigenvalue Problem
Find all the values of $\lambda$ for which the equation

\begin{equation*}
{\bf{A}}\boldsymbol{x} = \lambda \boldsymbol{x} \tag{1.34}
\end{equation*}

has a non-zero solution $\boldsymbol{x}$, where ${\bf{A}}$ is an $n\times n$ matrix and $\boldsymbol{x}$ is an $n\times 1$ matrix (column vector).

Such values of $\lambda$ are called **eigenvalues** of the matrix ${\bf{A}}$, and the corresponding vectors, $\boldsymbol{x}$, are called the **eigenvectors**.

:::



Equation (1.34) may be written as

\begin{eqnarray*}
&& \qquad\ \ {\bf{A}}\boldsymbol{x} &=& \lambda {\bf{I}}\boldsymbol{x} \\
\Leftrightarrow &\quad& {\bf{A}}\boldsymbol{x} - \lambda {\bf{I}}\boldsymbol{x} &=& {\bf{0}} \\
\Leftrightarrow && \left({\bf{A}} - \lambda {\bf{I}}\right)\boldsymbol{x} &=& {\bf{0}} \tag{1.35}
\end{eqnarray*}

The symbol "$\Leftrightarrow$" means "if and only if", *i.e.*, the rows in (1.35) are equivalent.

**If** $\textrm{det}\left({\bf{A}} - \lambda{\bf{I}}\right)\ne0$, the matrix ${\bf{A}} - \lambda{\bf{I}}$ may be inverted and the unique solution to (1.35) is

\begin{equation*}
\boldsymbol{x} = \left({\bf{A}} - \lambda{\bf{I}}\right)^{-1}{\bf{0}} = {\bf{0}}.
\end{equation*}

Non-zero solutions to (1.35) for $\boldsymbol{x}$ will only exist if

\begin{equation*}
\textrm{det}\left({\bf{A}} - \lambda{\bf{I}}\right) = 0. \tag{1.36}
\end{equation*}

There are **infinitely many** eigenvectors corresponding to any given eigenvalue.

This is because: if $\boldsymbol{x}$ is an eigenvector of ${\bf{A}}$ corresponding to the eigenvalue $\lambda$ and $c$ is a non-zero scalar, then $c\boldsymbol{x}$ is also an eigenvector of ${\bf{A}}$.  To see this note

\begin{equation*}
{\bf{A}}\left(c\boldsymbol{x}\right) = c{\bf{A}}\boldsymbol{x} = c\lambda\boldsymbol{x} = \lambda\left(c\boldsymbol{x}\right).
\end{equation*}

## Eigenvalues of $2\times2$ Matrices

If ${\bf{A}}$ is the $2\times 2$ matrix $\displaystyle {\bf{A}} = \left(\begin{array}{cc} a & b \\ c & d \end{array}\right)$ then (1.36) is a quadratic equation for $\lambda$.

We can see this by calculating

\begin{eqnarray*}
\left|{\bf{A}} - \lambda{\bf{I}}\right| &=& 
\left|\left(\begin{array}{cc} a & b \\ c & d \end{array}\right) - \lambda
\left(\begin{array}{cc} 1 & 0 \\ 0 & 1 \end{array}\right)\right| \\
&=& 
\left|\begin{array}{cc} a - \lambda & b \\ c & d - \lambda \end{array}\right| \\
&=& \left(a - \lambda\right)\left(d - \lambda\right) - bc \\
&=& \lambda^2 - \left(a + d\right)\lambda + \left(ad - bc\right).
\end{eqnarray*}

For $\lambda$ to be an eigenvalue of ${\bf{A}}$ we require

\begin{equation*}
\lambda^2 - \left(a + d\right)\lambda + \left(ad - bc\right) = 0.
\end{equation*}

We can factorize, complete the square, or use the standard formula for solving a quadratic equation to obtain the eigenvalues.

## The Characteristic Equation

:::{prf:definition} The Characteristic Equation

If ${\bf{A}}$ is an $n\times n$ matrix then $\textrm{det}\left({\bf{A}} - \lambda{\bf{I}}\right) $ is a polynomial of degree $n$ in $\lambda$, called the **characteristic polynomial**.  The **characteristic equation** is

\begin{equation*}
\textrm{characteristic polynomial} = 0,
\end{equation*}

*i.e.*,

\begin{equation*}
\lambda^n + c_{n-1}\lambda^{n-1} + c_{n-2}\lambda^{n-2} + \cdots + c_1\lambda + c_0 = 0. \tag{1.37}
\end{equation*}

:::

The coefficients $c_0$, $c_1$, $\ldots$, $c_{n-1}$ can be found once the elements of ${\bf{A}}$ are known.

The characteristic equation has $n$ roots (solutions of (1.37)) $\lambda_1$, $\lambda_2$, $\ldots$, $\lambda_n$.  Some of these roots may be equal, in which case they are said to be **repeated roots**.

:::{prf:example}

Find the eigenvalues and corresponding eigenvectors of the matrix

\begin{equation*}
{\bf{A}} = \left(\begin{array}{rr} 3 & -2 \\ 4 & -3 \end{array}\right).
\end{equation*}

:::{admonition} Solution

* **Step 1**  
Find the characteristic polynomial

\begin{eqnarray*}
\textrm{det}\left({\bf{A}} - \lambda{\bf{I}}\right) 
&=& \left|\begin{array}{cc} 3 - \lambda & -2 \\ 4 & -3 - \lambda \end{array}\right| \\
&=& \left(3 - \lambda\right)\left(-3 - \lambda\right) + 8 \\
&=& \lambda^2 - 9 + 8 = \lambda^2 - 1
\end{eqnarray*}

* **Step 2**  
Solve the characteristic equation

\begin{equation*}
\lambda^2 - 1 = 0
\quad\Rightarrow\quad
\lambda = -1~\textrm{or}~\lambda = 1.
\end{equation*}

* **Step 3**  
For **each** eigenvalue find the corresponding eigenvector.
    * Case $\lambda = -1$:  
    Write out  
    $\displaystyle \qquad\qquad\qquad
    \left({\bf{A}} - \lambda{\bf{I}}\right)\boldsymbol{x} = {\bf{0}}
    $    
    in full to give
    
    \begin{eqnarray*}
    &\left[\left(\begin{array}{rr} 3 & -2 \\ 4 & -3 \end{array}\right) + \left(\begin{array}{cc} 1 & 0 \\ 0 & 1
    \end{array}\right)\right]\left(\begin{array}{c} x_1 \\ x_2 \end{array}\right) &=& \left(\begin{array}{c} 0 \\ 0\end{array}\right)\\[2ex]
    \Rightarrow & \qquad\qquad\quad\ \left(\begin{array}{cc} 4 & -2 \\ 4 & -2 \end{array}\right)\left(\begin{array}{c} x_1 \\ x_2 \end{array}\right) &=& \left(\begin{array}{c} 0 \\ 0\end{array}\right)\\[2ex]
    \Rightarrow & \qquad\qquad\qquad\qquad 4x_1 - 2x_2 &=& 0.
    \end{eqnarray*}
    
    &emsp;&emsp;&nbsp;Hence any vector of the form $\displaystyle s\left(\begin{array}{c} 1 \\ 2\end{array}\right)$ is an eigenvector,
    
    &emsp;&emsp;&nbsp;(where $s\ne0$ is an arbitrary constant.)
    
    * Case $\lambda = 1$:  
    Write out  
    $\displaystyle \qquad\qquad\qquad
    \left({\bf{A}} - \lambda{\bf{I}}\right)\boldsymbol{x} = {\bf{0}}
    $    
    in full to give
    
    \begin{eqnarray*}
    &\left[\left(\begin{array}{rr} 3 & -2 \\ 4 & -3 \end{array}\right) - \left(\begin{array}{cc} 1 & 0 \\ 0 & 1
    \end{array}\right)\right]\left(\begin{array}{c} x_1 \\ x_2 \end{array}\right) &=& \left(\begin{array}{c} 0 \\ 0\end{array}\right)\\[2ex]
    \Rightarrow & \qquad\qquad\quad\ \left(\begin{array}{cc} 2 & -2 \\ 4 & -4 \end{array}\right)\left(\begin{array}{c} x_1 \\ x_2 \end{array}\right) &=& \left(\begin{array}{c} 0 \\ 0\end{array}\right)\\[2ex]
    \Rightarrow & \qquad\qquad\qquad\qquad x_1 - x_2 &=& 0.
    \end{eqnarray*}
    
    &emsp;&emsp;&nbsp;Hence any vector of the form $\displaystyle t\left(\begin{array}{c} 1 \\ 1\end{array}\right)$ is an eigenvector,
    
    &emsp;&emsp;&nbsp;(where $t\ne0$ is an arbitrary constant.)

# 3 x 3 Eigenvalues and Eigenvectors

Eigenvalue problems for $3\times3$ systems are of the same form as we have met already.

For a $3\times3$ matrix ${\bf{A}}$, we need to find eigenvalues $\lambda$ and eigenvectors $\boldsymbol{x}$ such that

\begin{equation*}
{\bf{A}}\boldsymbol{x} = \lambda\boldsymbol{x}.
\end{equation*}

In the case of a $3\times3$ system the characteristic polynomial will be a cubic and, in general, will have three roots and a corresponding number of eigenvectors.

:::{prf:example}

Find the eigenvalues and corresponding eigenvectors of the matrix

\begin{equation*}
{\bf{A}} = \left(\begin{array}{rrr} 1 & 2 & 1 \\ 0 & 1 & 2 \\ 1 & 0 & -3 \end{array}\right)
\end{equation*}

:::{admonition} Solution

* **Step 1**  
Find the characteristic polynomial

\begin{eqnarray*}
\textrm{det}\left({\bf{A}} - \lambda{\bf{I}}\right) &=& \left|\begin{array}{ccc}
1 - \lambda & 2 & 1 \\ 0 & 1 - \lambda & 2 \\ 1 & 0 & -3 - \lambda \end{array}\right| \\[2ex]
&=& \left|\begin{array}{cc} 2 & 1~ \\ 1 - \lambda & 2~ \end{array}\right| - \left(3 + \lambda\right)
\left|\begin{array}{cc}1 - \lambda & 2 \\0 & 1 - \lambda \end{array}\right|\quad\begin{array}{l}\textrm{Expanding by} \\ \textrm{the }3^{\textrm{rd}}\textrm{ row}\end{array}\\[1ex]
&=&\left[4 - \left(1 - \lambda\right)\right] - \left(3 + \lambda\right)\left[\left(1 - \lambda\right)^2\right]\\
&=&\left(3 + \lambda\right) - \left(3 + \lambda\right)\left[\left(1 - \lambda\right)^2\right]\\
&=&\left(3 + \lambda\right)\left[1 - \left(1 - \lambda\right)^2\right]\\
&=&\left(3 + \lambda\right)\left[2\lambda - \lambda^2\right]\\
&=&\left(3 + \lambda\right)\lambda\left(2 - \lambda\right).
\end{eqnarray*}

* **Step 2**  
Solve the characteristic equation  
$\displaystyle \qquad\qquad\qquad \textrm{det}\left({\bf{A}} - \lambda{\bf{I}}\right) =
\left(3 + \lambda\right)\lambda\left(2 - \lambda\right) = 0.$  
The solutions are

\begin{equation*}
\lambda = -3, \quad \lambda = 0, \quad \lambda = 2.
\end{equation*}

* **Step 3**. 
For each eigenvalue, find the corresponding eigenvector

    * Case $\lambda = -3$:  
    Write out  
    $\displaystyle \qquad\qquad\qquad
    \left({\bf{A}} - \lambda{\bf{I}}\right)\boldsymbol{x} = {\bf{0}}
    $    
    in full to give
    
    \begin{eqnarray*}
    &\left[\left(\begin{array}{rrr} 1 & 2 & 1 \\ 0 & 1 & 2 \\ 1 & 0 & -3 \end{array}\right) + 3\left(\begin{array}{ccc} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1
    \end{array}\right)\right]\left(\begin{array}{c} x_1 \\ x_2 \\ x_3 \end{array}\right) &=&\ \left(\begin{array}{c} 0 \\ 0 \\ 0\end{array}\right)\\[2ex]
    \Rightarrow & \qquad\qquad\qquad\qquad\quad \left(\begin{array}{ccc} 4 & 2 & 1 \\ 0 & 4 & 2 \\ 1 & 0 & 0 \end{array}\right)\left(\begin{array}{c} x_1 \\ x_2 \\ x_3 \end{array}\right) &=&\ \left(\begin{array}{c} 0 \\ 0 \\ 0\end{array}\right)\\[2ex]
    \Rightarrow & \qquad\qquad\qquad\qquad\qquad\qquad\  \left\{ \begin{array}{rcl} 4x_2 + 2x_3 &=& 0 \\ x_1 &=& 0\end{array} \!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\right.
    \end{eqnarray*}
    
    &emsp;&emsp;&nbsp;If these last two conditions are met we automatically have $4x_1 + 2x_2 + x_3=0$.
    
    &emsp;&emsp;&nbsp;Hence any vector of the form $\displaystyle r\left(\begin{array}{r} 0 \\ 1 \\ -2\end{array}\right)$ is an eigenvector.
    
    &emsp;&emsp;&nbsp;(where $r\ne0$ is an arbitrary constant.)
    
    * Case $\lambda = 0$:  
    Write out
    
    \begin{equation*}
    \left({\bf{A}} - \lambda{\bf{I}}\right)\boldsymbol{x} = {\bf{0}}
    \end{equation*}
    
    \begin{eqnarray*}
    &\left[\left(\begin{array}{rrr} 1 & 2 & 1 \\ 0 & 1 & 2 \\ 1 & 0 & -3 \end{array}\right) + 0\left(\begin{array}{ccc} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1
    \end{array}\right)\right]\left(\begin{array}{c} x_1 \\ x_2 \\ x_3 \end{array}\right) &=&\ \left(\begin{array}{c} 0 \\ 0 \\ 0\end{array}\right)\\[2ex]
    \Rightarrow & \qquad\qquad\qquad\qquad\quad \left(\begin{array}{rrr} 1 & 2 & 1 \\ 0 & 1 & 2 \\ 1 & 0 & -3 \end{array}\right)\left(\begin{array}{c} x_1 \\ x_2 \\ x_3 \end{array}\right) &=&\ \left(\begin{array}{c} 0 \\ 0 \\ 0\end{array}\right)\\[2ex]
    \Rightarrow & \qquad\qquad\qquad\qquad\qquad\qquad\quad  \left\{ \begin{array}{rcl} x_2 + 2x_3 &=& 0 \\ x_1 - 3x_3 &=& 0\end{array} \!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\right.
    \end{eqnarray*}
    
    &emsp;&emsp;&nbsp;If these last two conditions are met we automatically have $x_1 + 2x_2 + x_3=0$.
    
    &emsp;&emsp;&nbsp;Hence any vector of the form $\displaystyle s\left(\begin{array}{r} 3 \\ -2 \\ 1\end{array}\right)$ is an eigenvector.
    
    &emsp;&emsp;&nbsp;(where $s\ne0$ is an arbitrary constant.)
    
    * Case $\lambda = 2$:  
    Write out
    
    \begin{equation*}
    \left({\bf{A}} - \lambda{\bf{I}}\right)\boldsymbol{x} = {\bf{0}}
    \end{equation*}
    
    \begin{eqnarray*}
    &\left[\left(\begin{array}{rrr} 1 & 2 & 1 \\ 0 & 1 & 2 \\ 1 & 0 & -3 \end{array}\right) - 2\left(\begin{array}{ccc} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1
    \end{array}\right)\right]\left(\begin{array}{c} x_1 \\ x_2 \\ x_3 \end{array}\right) &=&\ \left(\begin{array}{c} 0 \\ 0 \\ 0\end{array}\right)\\[2ex]
    \Rightarrow & \qquad\qquad\qquad\qquad \left(\begin{array}{rrr} -1 & 2 & 1 \\ 0 & -1 & 2 \\ 1 & 0 & -5 \end{array}\right)\left(\begin{array}{c} x_1 \\ x_2 \\ x_3 \end{array}\right) &=&\ \left(\begin{array}{c} 0 \\ 0 \\ 0\end{array}\right)\\[2ex]
    \Rightarrow & \qquad\qquad\qquad\qquad\qquad\qquad\  \left\{ \begin{array}{rcl} -x_2 + 2x_3 &=& 0 \\ x_1 - 5x_3 &=& 0\end{array} \!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\!\right.
    \end{eqnarray*}
    
    &emsp;&emsp;&nbsp;If these last two conditions are met we automatically have $-x_1 + 2x_2 + x_3=0$.
    
    &emsp;&emsp;&nbsp;Hence any vector of the form $\displaystyle t\left(\begin{array}{r} 5 \\ 2 \\ 1\end{array}\right)$ is an eigenvector.
    
    &emsp;&emsp;&nbsp;(where $t\ne0$ is an arbitrary constant.)

# Properties of Determinants

The properties described below follow from the definition of the determinant (although proving them can be tricky for high-order matrices).

The properties apply to determinants of all orders, but are illustrated by $2\times2$ determinants

1. The determinant of any matrix is equal to the determinant of its transpose, *i.e.*, $\textrm{det}\left({\bf{A}}\right) = \textrm{det}\left({\bf{A}^\mathsf{T}}\right)$.  Therefore any results about determinants that deal with rows are also true for columns.

2. If all elements of one row of a matrix are multiplied by a constant $\lambda$, the determinant of the new matrix is $\lambda$ times the determinant of the original matrix.  For example

\begin{equation*}
\left|\begin{array}{rr} \lambda a & \lambda b \\ c & d \end{array} \right| = \lambda ad - \lambda bc
=\lambda\left(ad - bc\right) = \lambda \left|\begin{array}{rr} a & b \\ c & d \end{array} \right|
\end{equation*}

&emsp;&emsp;&nbsp;The same holds if one column is multiplied by a constant $\lambda$.

:::{admonition} Application

If a factor $\lambda$ appears in each element of a row or column of a matrix it can be taken out as a factor: *e.g.*

\begin{eqnarray*}
\left|\begin{array}{cc} 2 & 12 \\ 1 & 3 \end{array} \right| &=&
2\times\left|\begin{array}{rr} 1 & 6 \\ 1 & 3 \end{array} \right|=
2\times3\times\left|\begin{array}{rr} 1 & 2 \\ 1 & 1 \end{array} \right| \\[2ex]
&=& 2\times3\times\left(1\times 1 - 2\times1\right) = -6
\end{eqnarray*}

:::

:::{admonition} Application

If all elements of a row or column of a matrix are zero, the value of the determinant is zero.  For example

\begin{equation*}
\left|\begin{array}{cc} 0 & 0 \\ c & d \end{array} \right| = 0\times d - 0\times c = 0.
\end{equation*}

:::

:::{admonition} Application

If {\bf{A}} is an $n\times n$ matrix then

\begin{equation*}
\textrm{det}\left(\lambda{\bf{A}}\right) = \lambda^n \textrm{det}\left({\bf{A}}\right).
\end{equation*}

This result follows from applying rule 2 $n$ times, once for each row.

:::

3. Interchanging any two rows or any two columns of a determinant changes the sign of the determinant.  For example, interchanging rows gives

\begin{eqnarray*}
\left|\begin{array}{cc} c & d \\ a & b \end{array} \right| &=& cb - ad \\[2ex]
&=& -\left(ad - bc\right) \\[2ex]
&=& -\left|\begin{array}{cc} a & b \\ c & d \end{array} \right|
\end{eqnarray*}

:::{admonition} Application

If any two rows are identical, the determinant is zero.  Similarly, if any two columns are identical then the determinant is zero.

:::

:::{admonition} Application (of rules 2 and 3 together)

If any row is a multiple of any other row, the determinant is zero.  (Similarly if any column is a multiple of any other column, the determinant is zero.)  For example

\begin{equation*}
\left|\begin{array}{cc} \lambda c &  \lambda d \\ c & d \end{array} \right| =
\lambda \left|\begin{array}{cc} c &  d \\ c & d \end{array} \right| = \lambda\left(cd -cd\right) = 0.
\end{equation*}

:::

4. The value of a determinant is unchanged by adding to any row (or column) a constant multiple of any other row (or column).  For  example, if $\textrm{Row}~1 \rightarrow \left(\textrm{Row}~1 + \lambda\, \textrm{Row}~2\right)$,

\begin{eqnarray*}
\left|\begin{array}{cc} a + \lambda c &  b + \lambda d \\ c & d \end{array} \right| &=&
\left(a + \lambda c\right)d - \left(b + \lambda d\right)c \\[2ex]
&=& ad-bc = \left|\begin{array}{cc} a &  b \\ c & d \end{array} \right|.
\end{eqnarray*}

5. If ${\bf{A}}$ and ${\bf{B}}$ are square matrices of the same order then

\begin{equation*}
\textrm{det}\left({\bf{A}}{\bf{B}}\right) = \textrm{det}\left({\bf{A}}\right)\times\textrm{det}\left({\bf{B}}\right)
\end{equation*}

:::{admonition} Application

Let ${\bf{A}} = {\bf{I}}$ and let ${\bf{B}}$ be any matrix with non-zero determinant.  Then because ${\bf{A}}{\bf{B}} = {\bf{I}}{\bf{B}} = {\bf{B}}$, we have

\begin{equation*}
\textrm{det}\left({\bf{A}}{\bf{B}}\right) = \textrm{det}\left({\bf{B}}\right)
\end{equation*}
\begin{equation*}
=\textrm{det}\left({\bf{A}}\right)\textrm{det}\left({\bf{B}}\right) = \textrm{det}\left({\bf{I}}\right)\textrm{det}\left({\bf{B}}\right)
\end{equation*}

and therefore $\textrm{det}\left({\bf{I}}\right) = 1$.

:::

:::{prf:example} Determinant Calculation

Evaluate $\displaystyle \left|\begin{array}{rrr} 1 & -1 & 2 \\ 0 & 2 & 4 \\ 3 & 0 & 6 \end{array}\right|$

:::{admonition} Solution
:class: dropdown

\begin{eqnarray*}
\left|\begin{array}{rrr} 1 & -1 & 2 \\ 0 & 2 & 4 \\ 3 & 0 & 6 \end{array}\right| &=&
2\times \left|\begin{array}{rrr} 1 & -1 & 2 \\ 0 & 1 & 2 \\ 3 & 0 & 6 \end{array}\right| & \qquad\textrm{since Row 2 has a factor 2} \\[2ex]
&=&2\times3\times \left|\begin{array}{rrr} 1 & -1 & 2 \\ 0 & 1 & 2 \\ 1 & 0 & 2 \end{array}\right| & \qquad\textrm{since Row 3 has a factor 3} \\[2ex]
&=&2\times3\times \left|\begin{array}{rrr} 0 & -1 & 0 \\ 0 & 1 & 2 \\ 1 & 0 & 2 \end{array}\right| & \qquad\textrm{subtract Row 3 from Row 1} \\[2ex]
&=&2\times3\times \left|\begin{array}{cc} 0  & 2 \\ 1 &  2 \end{array}\right| & \qquad\textrm{expand by Row 1} \\[2ex]
&=&2\times3\times \left(0-2\right) & \\[2ex]
&=&-12.
\end{eqnarray*}

:::

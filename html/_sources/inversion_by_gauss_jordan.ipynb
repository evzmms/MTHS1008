# Inversion by the Gauss-Jordan Method

Row reduction methods can be used to find the inverse of a matrix

We will demonstrate matrix inversion "by Gauss-Jordan" by example

Only EROs may be applied: column operations are not allowed

:::{prf:example} 
Calculate the inverse of

\begin{equation*}
{\bf{A}} = \left(\begin{array}{ccc} 1 & 1 & 3 \\ 2 & 1 & 1 \\ 1 & 3 & 5 \end{array}\right).
\end{equation*}

* **Step 1**  
Consider the augmented matrix $\left[{\bf{A}}|{\bf{I}}\right]$  
$~$  
$ \displaystyle \qquad\qquad\qquad\qquad\qquad
\left[{\bf{A}}|{\bf{I}}\right]=
\left[\begin{array}{ccc|ccc} 
1 & 1 & 3 & 1 & 0 & 0 \\ 
2 & 1 & 1 & 0 & 1 & 0 \\
1 & 3 & 5 & 0 & 0 & 1
\end{array}\right].
$  
$~$  
The method is to perform elementary row operations on the augmented matrix to reduce it to the form $\left[{\bf{I}}|{\bf{B}}\right]$.  The $3\times3$ matrix ${\bf{B}}$ is then ${\bf{A}}^{-1}$, the inverse of ${\bf{A}}$. (Proof not required.)

* **Step 2**  
Clear the terms below the leading diagonal of the left-hand matrix

\begin{equation*}
\left[{\bf{A}}|{\bf{I}}\right]=
\left[\begin{array}{ccc|ccc} 
1 & 1 & 3 & 1 & 0 & 0 \\ 
\color{red}2 & 1 & 1 & 0 & 1 & 0 \\
\color{red}1 & \color{red}3 & 5 & 0 & 0 & 1
\end{array}\right]
\end{equation*}

\begin{equation*}
\left\{\begin{array}{l} 
\color{gray} R_1 \rightarrow R_1 \\
R_2 \rightarrow R_2 - 2R_1 \\
R_3 \rightarrow R_3 - R_1 
\end{array}\right.
\qquad
\left[\begin{array}{rrr|rrr} 
1 & 1 & 3 & 1 & 0 & 0 \\ 
\color{red}0 & -1 & -5 & -2 & 1 & 0 \\
\color{red}0 & 2 & 2 & -1 & 0 & 1
\end{array}\right]
\end{equation*}

\begin{equation*}
\left\{\begin{array}{l} 
\color{gray} R_1 \rightarrow R_1 \\
\color{gray} R_2 \rightarrow R_2 \\
R_3 \rightarrow R_3 + 2R_2 
\end{array}\right.
\qquad
\left[\begin{array}{rrr|rrr} 
1 & 1 & 3 & 1 & 0 & 0 \\ 
0 & -1 & -5 & -2 & 1 & 0 \\
0 & \color{red}0 & -8 & -5 & 2 & 1
\end{array}\right]
\end{equation*}

* **Step 3**  
Clear the terms above the leading diagonal of the left-hand matrix

\begin{equation*}
\left[\begin{array}{rrr|rrr} 
1 & \color{blue}1 & \color{blue}3 & 1 & 0 & 0 \\ 
0 & -1 & \color{blue}-5 & -2 & 1 & 0 \\
0 & 0 & -8 & -5 & 2 & 1
\end{array}\right]
\end{equation*}

\begin{equation*}
\left\{\begin{array}{l} 
R_1 \rightarrow 8R_1 + 3R_3 \\
R_2 \rightarrow 8R_2 - 5R_3 \\
\color{gray} R_3 \rightarrow R_3
\end{array}\right.
\qquad
\left[\begin{array}{rrr|rrr} 
8 & 8 & \color{blue}0 & -7 & 6 & 3 \\ 
0 & -8 & \color{blue}0 & 9 & -2 & -5 \\
0 & 0 & -8 & -5 & 2 & 1
\end{array}\right]
\end{equation*}

\begin{equation*}
\left\{\begin{array}{l} 
R_1 \rightarrow \phantom{8}R_1 + \phantom{3}R_2 \\
\color{gray} R_2 \rightarrow R_2 \\
\color{gray} R_3 \rightarrow R_3
\end{array}\right.
\qquad
\left[\begin{array}{rrr|rrr} 
8 & \color{blue}0 & 0 & 2 & 4 & -2 \\ 
0 & -8 & 0 & 9 & -2 & -5 \\
0 & 0 & -8 & -5 & 2 & 1
\end{array}\right]
\end{equation*}

* **Step 4**  
Multiply each row so that each element on the leading diagonal is 1

\begin{equation*}
\left[\begin{array}{rrr|rrr} 
\color{green}8 & 0 & 0 & 2 & 4 & -2 \\ 
0 & \color{green}-8 & 0 & 9 & -2 & -5 \\
0 & 0 & \color{green}-8 & -5 & 2 & 1
\end{array}\right]
\end{equation*}

\begin{equation*}
\left\{\begin{array}{l} 
R_1 \rightarrow R_1/8 \\
R_2 \rightarrow R_2/(-8) \\
R_3 \rightarrow R_3/(-8) \\
\end{array}\right.
\qquad
\left[\begin{array}{rrr|rrr} 
1 & 0 & 0 & \frac{1}{4} & \frac{1}{2} & -\frac{1}{4} \\[1ex] 
0 & 1 & 0 & -\frac{9}{8} & \frac{1}{4} & \frac{5}{8} \\[1ex]
0 & 0 & 1 & \frac{5}{8} & -\frac{1}{4} & -\frac{1}{8}
\end{array}\right]
\end{equation*}

* We have the augmented matrix in the form $\left[{\bf{I}}|{\bf{B}}\right]$.

* Hence, for $\displaystyle {\bf{A}} = \left(\begin{array}{ccc} 1 & 1 & 3 \\ 2 & 1 & 1 \\ 1 & 3 & 5 \end{array}\right)$ we have that $\displaystyle {\bf{A}}^{-1} = \left(\begin{array}{ccc}\frac{1}{4} & \frac{1}{2} & -\frac{1}{4} \\[1ex] -\frac{9}{8} & \frac{1}{4} & \frac{5}{8} \\[1ex]\frac{5}{8} & -\frac{1}{4} & -\frac{1}{8}
\end{array}\right)$.

:::{admonition} Exercise
Show that ${\bf{A}}{\bf{A}}^{-1} = {\bf{I}}$.
